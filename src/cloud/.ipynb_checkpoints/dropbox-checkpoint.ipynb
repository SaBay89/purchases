```python
import os
os.chdir('/root/projects/purchases/purchases/')
import dropbox
import logging
from dropbox.exceptions import AuthError, ApiError
from config.settings import DROPBOX_TOKEN
```

```python
def get_client():
    """Baut die Verbindung zu Dropbox auf und testet sie."""
    if not DROPBOX_TOKEN:
        logging.error("Kein DROPBOX_TOKEN gefunden!")
        return None
        
    try:
        dbx = dropbox.Dropbox(DROPBOX_TOKEN)
        # Kurzer Test, ob der Token gültig ist
        dbx.users_get_current_account()
        return dbx
    except AuthError as e:
        logging.error(f"Dropbox Authentifizierungsfehler (Token abgelaufen?): {e}")
        return None
```

```python
LOCAL_DOWNLOAD_DIR = "data/temp_images"
os.makedirs(LOCAL_DOWNLOAD_DIR, exist_ok=True)
```

```python
dbx = get_client()
```
